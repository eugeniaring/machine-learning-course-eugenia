### One-hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

cols_to_encode = ['job','marital']

df_encode = df[cols_to_encode]

encoder = OneHotEncoder()

encoded_data = encoder.fit_transform(df_encode)
encoded_data
```
